### Typical 1. Total Time to Watch a Series

Compute total viewing time (no breaks) for a series:
- 5 seasons, 10 episodes per season
- Season 1: 45 minutes 35 seconds per episode
- Seasons 2–5: 35 minutes 15 seconds per episode
Display the total as hours, minutes, and seconds.

**Example Output:**
```
Total time: 39h 30m 0s
```


````python
# SEASON 1 EPISODES IS 2735 SECONDS LONG
s1_ep = 2735

# SEASON 2-5 EPISODES IS 2115 SECONDS LONG
s2_to_5_ep = 2115

total_time = 0
total_time = s1_ep * 10 # THERE ARE 10 EPISODES IN SEASON 1
total_time = total_time + (s2_to_5_ep *40) # THERE ARE TOTAL 40 EPISODES STARTING FROM SEASON 2 TO SEASON 5

# THE REMAINDER OF THIS DIVISION GIVES THE AMOUNT OF SECONDS THAT IS LESS THAN A MINUTE
seconds = total_time % 60

# THIS SUBTRACTION WILL GIVE THE TOTAL NUMBER OF MINUTES
total_minutes = total_time - seconds
total_minutes = total_minutes/60 # SECONDS CONVERTED INTO TOTAL MINUTES

# THE REMAINDER OF THIS DIVISION GIVES THE AMOUNT OF MINUTES THAT IS LESS THAN A HOUR
remaining_min = int(total_minutes %60)

# THIS SUBTRACTION WILL GIVE THE TOTAL NUMBER OF HOURS IN MINUTES FORM
mins = total_minutes - remaining_min

# THE NUMBER OF HOURS THAT WERE IN MINUTES FORM ARE NOW CONVERTED IN HOURS FORM
hours = int(mins/60)

#DISPLAYING THE TOTAL NUMBER OF HOURS, MINUTES AND SECONDS THAT WILL TAKE TO FINISH THE SERIES
print("Total Time taken is: "+str(hours)+"h "+str(remaining_min)+"m "+str(seconds)+"s ")
````

### Stretch 1. Sum of First n Integers

Ask for a positive integer `n` and compute the sum `n*(n+1)//2`. Print the result.

**Example Output:**
```
Sum of 1..10 = 55
```


````python
sum = 0

# ASKING USER TO INPUT 
num = int(input("Enter a Postive Integer: \n"))

# THIS IS TO MAKE AN ENDLESS LOOP IF THE CONDITION IS NOT SATISIFIED. CONDITION IS TO ENTER A POSITIVE INTEGER
while num <= 0:
    num = int(input("Enter a Postive Integer: \n"))

# I HAVE INCREMENTED NEW'S VALUE SO THE FOR LOOP CAN RUN N TIMES.
new = num + 1
for x in range(new):
    sum = sum + x

# THIS IS TO DISPLAY SUM OF n*(n+1)//2
print("Sum of 1 to",num,"is =",sum)
````

### Stretch 3. Population Projection (5 Years)

Assume current population is 312032486 and 1 year = 365 days.
Rates: birth every 7s, death every 13s, immigrant every 45s.
Compute projected population for each of the next 5 years using integer division (`//`).

**Example Output:**
```
Year 1: 315,xxx,xxx
...
Year 5: ...
```


*Optional hint:* Compute yearly net change once; then add each year.


````python
# STUDENT NAME: ALI ASAD
# MISIS NUMBER: M01091335

# CURRENT POPULATION VALUE
current_population = 312032486

# TO CACULATE 365 DAYS (A YEAR) IN SECONDS
year_into_seconds = 365 * 60 * 60 * 24

# TO CACULATE HOW MANY BIRTHS, DEATHS AND IMMIGRANTS ARE THERE IN A YEAR
total_births = year_into_seconds // 7 
total_deaths = year_into_seconds // 13
total_immigrants = year_into_seconds // 45

# THIS IS TO FIND THE TOTAL VALUE OF POPULATION INCREASED OR DECREASED
sum = (total_births + total_immigrants) - total_deaths

# THIS IS TO DISPLAY EVERY YEAR'S POPULATION PROJECTION
for x in range(5):
    x = x+1
    current_population = current_population + sum
    print("Year",x,": ",current_population)
````

### Stretch 4. Payroll Statement

Ask for: employee name, hours worked, hourly rate, federal tax rate, state tax rate.
Print a formatted payroll slip with gross pay, each deduction, and net pay (2 d.p.).

**Example Output:**
```
Employee: Smith
Gross pay: £97.50
Federal withholding: £19.50
State withholding: £8.78
Net pay: £69.22
```


````python
# ASKING USER TO INPUT THE FOLLOWING INFORMATION: EMPLOYEE'S NAME, HOW MANY HOURS WORKED, THE HOURLY RATE, THE FEDERAL TAX RATE & STATE TAX RATE
name = input("Enter Employee's name: \n")
hours = int(input("how many hours Worked: \n"))
hourly_rate = float(input("What is the hourly rate: \n"))
federal_tax_rate = float(input("What is the Federal Tax Rate: \n"))
state_tax_rate = float(input("What is the State Tax Rate: \n"))

# TO CACULATE THE FOLLOWING VALUES: GROSS PAY
gross_pay = (hours * hourly_rate)

# THIS IS TO CALCULATE THE FEDERAL TAX AND STATE TAX AMOUNT
federal_tax = (gross_pay*(federal_tax_rate/100))
state_tax = (gross_pay*(state_tax_rate/100))

# THIS IS TO CALCULATE THE NET PAY
net_pay = round((gross_pay - federal_tax - state_tax),2)


# THIS IS TO DISPLAY THE EMPLOYEE NAME, GROSS PAY, FEDERAL WITHHOLDING, STATE WITHHOLDING AND THE NET PAY
print("Employee Name:", name)
print("Gross Pay: £"+str(gross_pay))
print("Federal withholding: £"+str(federal_tax))
print("State withholding: £"+str(state_tax))
print("Net pay: £"+str(net_pay))
````
