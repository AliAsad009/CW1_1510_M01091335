4. **Body mass index (BMI)** is a measure of health and weight.

It can be calculated by taking your weight in kilograms and dividing it by the square of your height in meters. The interpretation of BMI for people 20 years or older is as follows:


BMI	Interpretation
BMI < 18.5	Underweight
18.5 ≤ BMI ≤ 25.0	Normal
25.0 ≤ BMI ≤ 30.0	Overweight
30.0 > BMI 	Obese
 
Write a program that prompts the user to enter weight in pounds and height in inches and display the BMI. Note that one pound is 0.45359237 kilograms, and one inch is 0.0254 meters


````python

# ASKS USER TO INPUT WEIGHT IN POUNDS AND HEIGHT VALUES IN INCHES RESPECTIVELY
weight = float(input("Enter your weight in Pounds:\n"))
height = float(input("Enter your height in Inches:\n"))

inch_to_m = height * 0.0254 # THIS IS THE FORMULA TO CONVERT HEIGHT FROM INCH TO METRES
pounds_to_kgs = weight * 0.45359237 # THIS IS THE FORMULA TO CONVERT WEIGHT FROM POUNDS TO KGS

BMI = pounds_to_kgs / ((inch_to_m)**2) # THIS IS THE FORMULA TO CALCULATE BMI

if BMI < 18.5: # THIS CONDITION CHECKS IF BMI IS LESS THAN 18.5
    print("Underweight") # IF BMI IS LESS THAN 18.5, IT PRINTS UNDERWEIGHT
elif BMI >= 18.5 and BMI < 25.0: # THIS CONDITION CHECKS IF BMI IS GREATER THAN OR EQUAL TO 18.5 AND LESS THAN 25.0
    print("Normal") # IF BMI IS GREATER THAN OR EQUAL TO 18.5 AND LESS THAN 25 IT PRINTS NORMAL
elif BMI >= 25.0 and BMI < 30.0: # THIS CONDITION CHECKS IF BMI IS GREATER THAN OR EQUAL TO 25.0 AND LESS THAN 30.0
    print("Overweight") # IF BMI IS GREATER THAN OR EQUAL TO 25.0 AND LESS THAN 30.0 IT PRINTS OVERWEIGHT
else: # IF NO CONDITION ABOVE IS MET, THIS RUNS
    print("Obese") # PRINTS OBESE IF NO ABOVE CONDITIONS IS MET

````

5. **Student Marks Analysis**
You are working as a data analyst in a school. You have the exam marks of students stored in a list. Your task is to analyze the data.

**Requirements**:

- Store marks of students in a list.

- Use a loop to calculate the average mark.

- Find the highest and lowest mark using list functions.

- Print the number of students who passed (marks ≥ 50) and failed.

````python
marks = [50,25,10]
total = 0
count = 0
passed = 0

# THIS FOR LOOP HELPS US IN ACCESING EACH NUMBER IN THE LIST
for num in marks:
    total = total + num # ADDS THE NUMBER IN THE LIST
    count = count + 1 # COUNT INCREMENTS
    if num >= 50: # CHECKS IF ANY NUMBER IN THE MARKS LIST IS GREATER THAN OR EQUAL TO 50
        passed = passed + 1 # IF ANY NUMBER IN THE LIST IS GREATER THAN OR EQUAL TO 50, IT INCREMENTS PASSED
        
average = total / count # FINDS THE AVERAGE MARK
print("Average is", average) # PRINTS AVERAGE MARK

max_num = max(marks) # FINDS THE HIGHEST MARKS
print("Highest Number is", max_num) # PRINTS THE HIGHEST MARKS

min_num = min(marks) # FINDS THE LOWEST MARKS
print("Lowest Number is",min_num) # PRINTS THE LOWEST MARKS

failed = len(marks) - passed
print(f'{passed} Students have passed. {failed} Students have failed') # PRINTS HOW MANY STUDENTS HAVE PASSED AND HOW MANY STUDENTS HAVE FAILED
````

16. Write a **guessing game** in Python where the user has to guess a secret number. After every guess the program tells the user whether their number was too large or too small. At the end the number of tries needed should be printed.


````python
import random # IMPORTS RANDOM MODULE
secret = random.randint(1,100) # THIS GENERATE A RANDOM 3 DIGIT NUMBER
num_of_tries = 0
check = False

while check == False:
    guess = int(input("\nEnter your guess:\n")) # ASKS THE USER TO INPUT THEIR GUESS
    if guess == secret: # CHECKS IF THE GUESS IS SAME AS THE SECRET NUMBER
        check = True
    else:
        num_of_tries = num_of_tries + 1 # IF THE CONDITION IS FALSE, NUM_OF_TRIES IS INCREMENTED
        if guess > secret: # IF GUESS WAS LARGER THAN SECRET
            print("Guess was too large") # IT WILL PRINT GUESS WAS TOO LARGE
        if guess < secret: # IF GUESS WAS SMALLER THAN SECRET
            print("Guess was too small") # IT WILL PRINT GUESS WAS TOO SMALL
    
print(f'It took you {num_of_tries} tries') # THIS WILL PRINT HOW MUCH TRIES IT TOOK YOU TO GUESS
````

**Requirements:**

2. Store all entered numbers in a **list**.
3. After the input ends:

   * If numbers were entered:

     * Print all the numbers entered.
     * Check the list and print the **first even number**. If no even number exists, print `"No even numbers found"`.
   * If no numbers were entered, print `"No numbers were entered"`.


> ### Note 
>
>* **`lower()`**: Converts any text into lowercase.
>  Example: `"DONE".lower()` becomes `"done"`.
>  This makes it easier to compare user input without worrying about uppercase/lowercase.
>
>* **`isdigit()`**: Checks if a string only contains digits (0–9).
>  Example: `"123".isdigit()` returns `True`, but `"12a".isdigit()` returns `False`.
>  This helps us make sure the user typed a valid number before converting it into an integer.
> 


---

````python
list = [] # GENERATES AN EMPTY LIST
end = False
evenfound = False
even = 0
count = 0

while end == False:
    answer = input("Enter a positive number, or type \"Done\" to stop:\n")
    if answer.lower() == "done": # CHECKS IF THE LOWERCASE OF THE USER INPUTTED VALUE IS SAME AS "done"
        end = True # IF THE CONDITION IS TRUE, END VALUE IS UPDATED TO TRUE
    elif float(answer) > 0: # CHECKS IF THE ANSWER IS GREATER THAN 0
        list.append(answer) # IF CONDITION IS TRUE, A NEW VALUE IS ADDED TO THE LIST
        count = count + 1 # COUNT IS INCREMENTED BY 1 TO SHOW HOW MANY NUMBERS ARE PRESENT IN THE LIST
        while evenfound == False: # THIS CONDITION WILL STOP WHEN A EVEN NUMBER IS FOUND
            if float(answer) % 2 == 0: # CHECKS IF THE ANSWER DIVIDED BY 2, LEAVES ANY REMAINDER
                even = answer # IF CONDITON IS TRUE, EVEN VALUE IS UPDATED
                evenfound = True # EVENFOUND VALUE IS UPDATED ONCE AN EVEN NUMBER IS FOUND
    

if count == 0: # CHECKS IF ANY NUMBER IS PRESENT IN THE LIST
    print("No Numbers were entered") # IF NO, IT PRINTS "No Numbers were entered"
else:
    print(list) # IF CONDITON IS FALSE, IT PRINTS THE LIST
    if evenfound == False: # CHECKS IF ONE EVEN NUMBER WAS FOUND
        print("No even numbers were found") # IF NO EVEN NUMBER IS FOUND, IT PRINTS "No even numbers were found"
    else:
        print(f'First even number found was {even}') # IF AN EVEN NUMBER IS FOUND, IT PRINTS THE EVEN NUMBER
````
