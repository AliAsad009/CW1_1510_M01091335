4. Create a function (or set of functions ) taking one parameter for a password.

The function (s) will print True if

  - the password contains at least one upper case letter,   
  - the password contains at least one lower case letter,
  - the password contains at least one digit and one special character like for example ‘@’.
- Extra (check if the password contains space. If so, print a suitable error message)

````python
def password_check(password):
    lowercase_count = 0
    uppercase_count = 0
    specialchar_count = 0
    digit_count = 0

    for x in password:

        if ord(x) == 32: # checks for space character
            print("Error! Password cannot contain [Space]. Try again!")
            return
            
        elif ord(x) >= 97 and ord(x) <= 122: # checks for lowercase characters
            lowercase_count += 1
            
        elif ord(x) >= 65 and ord(x) <= 90:   # checks for uppercase characters             
            uppercase_count += 1
            
        elif ord(x) >= 33 and ord(x) <= 47: #checks for special characters
            specialchar_count += 1
        elif ord(x) >= 58 and ord(x) <= 64: #checks for special characters
            specialchar_count += 1
        elif ord(x) >= 91 and ord(x) <= 96: #checks for special characters
            specialchar_count += 1
        elif ord(x) >= 123 and ord(x) <= 126: #checks for special characters
            specialchar_count += 1
       
        elif ord(x) >= 48 and ord(x) <= 57: #checks for digits
            digit_count += 1

    if lowercase_count == 0 or uppercase_count == 0 or specialchar_count == 0 or digit_count == 0: # checks if the password does not contain digit , special character , lower case or uppercase 
        print("The password should contain atleast one upper case letter, one lower case letter, one special character and one digit")
    else:
        print("True")
````

7. Create a function generating an email for an organization: Ask the user for his or her initials. Return the new email.

````python
def gen_email(): 
    import random
    initial = input("Please enter your initials without any space:   ")
    for x in initial:
        if x == " ": # this will check for space
            return False
    new = initial.lower() # will convert the inital in lowercase
    num = random.randint(100,999) # one random 3-digit number is generated
    print(f"{new}{num}@live.mdx.ac.uk") # concatenate the initials and the 3 digit number with @live.mdx.ac.uk

gen_email()
````

12. Password Validation with Multiple Criteria: Ask the user for a password and validate whether it:
- Has more than 8 characters.
- Contains at least one uppercase letter.
- Contains at least one lowercase letter.
- Contains at least one number.
- Contains at least one special character (@, #, $, etc.).
- Extra: If the password contains a space, print an error message

````python
def check_pw():
    upper_count = 0
    lower_count = 0
    digit_count = 0
    special_count = 0

    
    password = input("Enter your password:    ")
    if len(password) > 8:
        for x in password:
            if ord(x) >= 65 and ord(x) <= 90: # if ascii value of x is equal to or greater than 65 and less than or equal to 90
                upper_count += 1
            elif ord(x) >= 97 and ord(x) <= 122: # if ascii value of x is equal to or greater than 97 and less than or equal to 122
                lower_count += 1
            elif ord(x) >= 48 and ord(x) <= 57: # if ascii value of x is equal to or greater than 48 and less than or equal to 57
                digit_count += 1
            elif ord(x) >= 33 and ord(x) <= 47: # if ascii value of x is equal to or greater than 33 and less than or equal to 47
                special_count += 1
            elif ord(x) >= 58 and ord(x) <= 64: # if ascii value of x is equal to or greater than 58 and less than or equal to 64
                special_count += 1
            elif ord(x) >= 92 and ord(x) <= 96: # if ascii value of x is equal to or greater than 92 and less than or equal to 96
                special_count += 1
            elif ord(x) >= 123 and ord(x) <= 126: # if ascii value of x is equal to or greater than 123 and less than or equal to 126
                special_count += 1
            elif ord(x) == 32: # if ascii value of x is equal to 32
                print("Error. Cannot Contain [Space]")
                return False
        if special_count < 1 or lower_count < 1 or digit_count < 1 or special_count < 1: # will check if the password contains a lowercase, a uppercase, a digit, and a special character
            print("Error. Password must contain at least one uppercase letter, one lowercase letter, one number and one special character")
        else:
            print("Password meets all of the conditions")
    else:
        print("Password should have more than 8 characters") 

check_pw()
````

````python
# STUDENT NAME : ALI ASAD
# MISIS NUMBER : M01091335
def binaryToDecimal(binaryString): # uses binaryString as parameter
    sum = 0
    binaryString = binaryString[::-1] # list is reversed
    for x in range(len(binaryString)):
        if binaryString[x] == "1": # checks for 1 in the list
            value = (2**x) # value of the number x is the 2 power x
            sum += value # value is added in sum
    print(sum)

binaryToDecimal("10001")
````
