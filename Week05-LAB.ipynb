### Activity 13: Identifying Duplicate Entries in Clothing Stock

Create a program that reads the `clothing_stock.txt` file and identifies if there are any duplicate entries (both item and size). Use a set to find and print these duplicates.

**Steps:**
1. Read the file and store each item in a set.
2. Keep track of duplicate entries.
3. Print any duplicates found.
<hr>

`````python
with open("clothing_stock.txt", mode = "r") as f: # opens clothing_stock.txt in read mode
    clothing = f.readlines()
    y = set(clothing) # converts y into a set

    for x in y:
        count = 0
        for cloth in clothing:
            # This Checks for duplicates
            if x == cloth: 
                count += 1 # if both values are same, increment the value of count by 1
        if count > 1: 
            print(f'{count} Duplicates of {x.strip()} are found')
`````
